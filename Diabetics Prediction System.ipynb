***Input & Output Separation***

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x
y

```

***Train-Test Splitting***

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

***Normalization***

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

***Model Creation and Performance Evaluation***
        
Algorithms:

- K-Nearest Neighbors

- Naive Bayers

- Support Vector Machine

- Decision Tree

- Random Forest

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
knn=KNeighborsClassifier(n_neighbors=3)
nb=GaussianNB()
svm=SVC()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier(n_estimators=50)
lst=[knn,nb,svm,dt,rf]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
for i in lst:
  print(i)
  print('*********************************************')
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  # print('confusion metrics=',confusion_matrix(y_test,y_pred))
  print(classification_report(y_test,y_pred))
  labels=[0,1]
  result=confusion_matrix(y_test,y_pred)
  cmd=ConfusionMatrixDisplay(result,display_labels=labels)
  cmd.plot()
```

```python
new_labels=['KNN','NB','SVC','DT','RF']
colors=['black','purple','red','gray','green']
accuracy=[]
for i in lst:
  y_pred=i.predict(x_test)
  accuracy_score=classification_report(y_test,y_pred,output_dict=True)['accuracy']
  accuracy.append(accuracy_score)
plt.bar(new_labels,accuracy,color=colors)
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title=('Accuracy Score for Different Classifier')

for i in range (len(accuracy)):
  plt.text(i,accuracy[i],str(round(accuracy[i]*100,2))+'%',ha='center',va='bottom')
  # plt.show()
```

***Model Prediction***

```python
input_data=nb.predict(scaler.transform([[2,147,70,38,362,29,0.331,30]]))
if input_data==0:
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```

***Conclusion: Gaussian Naive Bayers as the optimal model***

The result of the five classifiers tested(KNeighborsClassifier,GaussianNB,SVC,DecisionTreeClassifier and RandomForestClassifier) indicates that Gaussian Naive Bayers emergesas the best performing model for the given data set as it give the highest accuracy of 79% comparing to the remaing classifier models.
